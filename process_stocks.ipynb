# Notebook to pre-process stocks.csv

```python
import pandas as pd
import numpy as np
#pd.set_option('display.max_rows', None)
#pd.set_option('display.max_columns', None)

stocks = pd.read_csv('dataset/historical_stocks.csv')
print(stocks.shape)
```

```python
stocks
```

```python
print("Unique values for stocks.csv")
for column in stocks:
    print(column, "unique values: ",stocks[column].nunique())
```

```python
sector = stocks.groupby(['name'])['sector'].nunique()
df = pd.DataFrame(sector)
count = [0,0,0]
names_to_delete = set()
for row in df.itertuples():
    if row.sector == 0:
        count[0] += 1
        names_to_delete.add(row.Index)
    if row.sector == 1:
        count[1] += 1
    if row.sector == 2:
        count[2] += 1
print("0 nunique \t 1 nunique \t 2 nunqique")
print(count[0],'\t\t', count[1], '\t\t', count[2])
```

```python
industry = stocks.groupby(['name'])['industry'].nunique()
df = pd.DataFrame(industry)
count = [0,0,0]
for row in df.itertuples():
    if row.industry == 0:
        count[0] += 1
        names_to_delete.add(row.Index)
    if row.industry == 1:
        count[1] += 1
    if row.industry == 2:
        count[2] += 1    
print("0 nunique \t 1 nunique \t 2 nunqique")
print(count[0],'\t\t', count[1], '\t\t', count[2])
```

```python
for s in names_to_delete:
    stocks = stocks[stocks.name != s]  
```
