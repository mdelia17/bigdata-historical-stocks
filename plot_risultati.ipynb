```python
#!pip install matplotlib==3.4.2 --user
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_tempi_esecuzione(data,bounds,filepath):

    labels = ["MapReduce","Hive","Spark"]

    x = np.arange(len(labels))  # the label locations
    width = 0.30  # the width of the bars

    fig, ax = plt.subplots(figsize=(9,5))
    rects1 = ax.bar(x, data[0], width, label='1M')
    rects2 = ax.bar(x + width, data[1], width, label='10M')
    rects3 = ax.bar(x + 2 * width, data[2], width, label='20M')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Secondi', labelpad=12, fontsize=14)
    ax.set_xticks(x + width)
    ax.set_xticklabels(labels, fontsize=14)
    ax.legend(fontsize=14, title="Numero righe", title_fontsize=12, loc=0)

    ax.bar_label(rects1, padding=2, fontsize=11)
    ax.bar_label(rects2, padding=2, fontsize=11)
    ax.bar_label(rects3, padding=2, fontsize=11)

    fig.tight_layout()

    plt.ylim(bounds)

    plt.show()

    fig.savefig(filepath)
```

```python
bounds_cluster = [0,800]
bounds_standalone = [0,1600]

# formato dati [[1M_MR, 1M_H, 1M_S],[10M_MR, 10M_H, 10M_S],[20M_MR, 20M_H, 20M_S]]
job1_cluster = [[100, 37, 42],[369, 138, 136],[722, 185, 255]]
job2_cluster = [[100, 51, 43],[335, 159, 151],[423, 220, 199]]
job3_cluster = [[176, 233, 47],[226, 525, 115],[286, 562, 200]]

job1_standalone = [[72, 226, 48],[570, 740, 454],[1353, 1266, 1027]]
job2_standalone = [[68, 242, 67],[424, 503, 590],[900, 795, 857]]
job3_standalone = [[260, 369, 84],[407, 739, 585],[1048, 940, 1395]]

plot_tempi_esecuzione(job1_cluster,bounds_cluster,"plot/job1_cluster.png")
plot_tempi_esecuzione(job2_cluster,bounds_cluster,"plot/job2_cluster.png")
plot_tempi_esecuzione(job3_cluster,bounds_cluster,"plot/job3_cluster.png")

plot_tempi_esecuzione(job1_standalone,bounds_standalone,"plot/job1_standalone.png")
plot_tempi_esecuzione(job2_standalone,bounds_standalone,"plot/job2_standalone.png")
plot_tempi_esecuzione(job3_standalone,bounds_standalone,"plot/job3_standalone.png")
```
